# Dataset Overview: Mediterranean Seagrass Presence-Absence Analysis

This notebook documents the preprocessing and analysis of Mediterranean seagrass distribution data, based on the approach from Effrosynidis et al. (2018). The dataset contains seagrass presence observations paired with pseudo-absence samples, each annotated with comprehensive environmental predictors.

## Data Files
- **`data/presence.txt`**: Field observations of seagrass occurrences with environmental covariates
- **`data/absence.txt`**: Algorithmically generated pseudo-absence samples with the same predictors
- **Format**: Tab-separated values (TSV) with identical column structure

## Key Insights from Variable Ranges

1. **Geographic Coverage**: The dataset spans the western Mediterranean (Spain) to the eastern regions (Italy/France), with latitude range typical for Mediterranean coastal zones.

2. **Temperature Gradients**: Clear seasonal variation (10.58-28.36°C) reflecting Mediterranean climate, with cooler deeper waters.

3. **Salinity Patterns**: Relatively stable salinity (35.97-38.79 PSU) typical of Mediterranean waters, with slight depth variations.

4. **Chlorophyll-a**: Low to moderate values (0.06-1.33 mg m⁻³) indicating oligotrophic to mesotrophic conditions.

5. **Nutrient Levels**: Generally low nitrate and phosphate concentrations, characteristic of Mediterranean oligotrophy.

6. **Coastal Proximity**: Most samples are very close to shore (0.0001-0.025 km), appropriate for seagrass habitat modeling.

## Data Quality Notes

- **Pseudo-absences**: The absence dataset was algorithmically generated, not empirically observed
- **Temporal Scope**: All environmental data from 2015; seasonal patterns captured through monthly aggregations  
- **Depth Layers**: `_maxDepth` variables provide deeper water column information where available
- **Missing Values**: Check for NaN values in preprocessing steps below

```python
# Libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load and analyze the dataset to compute exact variable ranges
# Set up data paths
DATA_DIR = Path(r"c:/Users/Guillem/Documents/GitHub/med_seagrass/data")
pres_file = DATA_DIR / "presence.txt"
abs_file = DATA_DIR / "absence.txt"

# Load the datasets
print("Loading datasets...")
df_pres = pd.read_csv(pres_file, sep="\t")
df_abs = pd.read_csv(abs_file, sep="\t")

print(f"Presence records: {len(df_pres):,}")
print(f"Absence records: {len(df_abs):,}")
print(f"Total records: {len(df_pres) + len(df_abs):,}")

# Convert numeric columns
numeric_cols = []
categorical_cols = ["BIO_CLASS", "Substrate"]

for col in pres.columns:
    if col not in categorical_cols:
        pres[col] = pd.to_numeric(pres[col], errors='coerce')
        absd[col] = pd.to_numeric(absd[col], errors='coerce')
        numeric_cols.append(col)

# Combine datasets for range computation
combined = pd.concat([pres, absd], ignore_index=True)

# Compute ranges for all numeric variables
print("\nComputing variable ranges...")
ranges = {}
for col in numeric_cols:
    if col in combined.columns:
        col_data = combined[col].dropna()
        if len(col_data) > 0:
            ranges[col] = {
                'min': col_data.min(),
                'max': col_data.max(),
                'mean': col_data.mean(),
                'std': col_data.std()
            }

# Display key ranges for verification
print("\n=== KEY VARIABLE RANGES ===")
key_vars = ['LONGITUDE', 'LATITUDE', 'VOTEMPER_2015_year', 'VOSALINE_2015_year', 
           'CHL_2015_year', 'NIT_2015_year', 'PHO_2015_year', 'ZSD_2015_year', 
           'VHM0_2015_year', 'Med_bathym', 'Distance_to_Coast']

for var in key_vars:
    if var in ranges:
        r = ranges[var]
        print(f"{var:25s}: {r['min']:8.4f} - {r['max']:8.4f} (mean: {r['mean']:6.4f})")

# Check categorical variables
print("\n=== CATEGORICAL VARIABLES ===")
print("BIO_CLASS unique values:")
print(combined['BIO_CLASS'].value_counts())
print(f"\nSubstrate unique values ({combined['Substrate'].nunique()} total):")
print(combined['Substrate'].value_counts().head(10))

# Save detailed ranges to file
ranges_df = pd.DataFrame(ranges).T
ranges_df = ranges_df.round(6)
ranges_df.to_csv(DATA_DIR / "variable_ranges_detailed.csv")
print(f"\nDetailed ranges saved to: {DATA_DIR / 'variable_ranges_detailed.csv'}")
```

```python
# Merge presence and absence dataframes into a single dataframe, add a new column 'Presence' to indicate presence (1) or absence (0) and save to a new CSV file
df_pres['Presence'] = 1
df_abs['Presence'] = 0
combined = pd.concat([df_pres, df_abs], ignore_index=True)

combined.to_csv(DATA_DIR / "combined_data.csv", index=False)
```
