# Dataset Overview: Mediterranean Seagrass Presence-Absence Analysis

This notebook documents the preprocessing and analysis of Mediterranean seagrass distribution data, based on the approach from Effrosynidis et al. (2018). The dataset contains seagrass presence observations paired with pseudo-absence samples, each annotated with comprehensive environmental predictors.

## Data Files
- **`data/presence.txt`**: Field observations of seagrass occurrences with environmental covariates
- **`data/absence.txt`**: Algorithmically generated pseudo-absence samples with the same predictors
- **Format**: Tab-separated values (TSV) with identical column structure

## Key Insights from Variable Ranges

1. **Geographic Coverage**: The dataset spans the western Mediterranean (Spain) to the eastern regions (Italy/France), with latitude range typical for Mediterranean coastal zones.

2. **Temperature Gradients**: Clear seasonal variation (10.58-28.36°C) reflecting Mediterranean climate, with cooler deeper waters.

3. **Salinity Patterns**: Relatively stable salinity (35.97-38.79 PSU) typical of Mediterranean waters, with slight depth variations.

4. **Chlorophyll-a**: Low to moderate values (0.06-1.33 mg m⁻³) indicating oligotrophic to mesotrophic conditions.

5. **Nutrient Levels**: Generally low nitrate and phosphate concentrations, characteristic of Mediterranean oligotrophy.

6. **Coastal Proximity**: Most samples are very close to shore (0.0001-0.025 degrees), appropriate for seagrass habitat modeling.

## Data Quality Notes

- **Pseudo-absences**: The absence dataset was algorithmically generated, not empirically observed
- **Temporal Scope**: All environmental data from 2015; seasonal patterns captured through monthly aggregations  
- **Depth Layers**: `_maxDepth` variables provide deeper water column information where available
- **Missing Values**: Check for NaN values in preprocessing steps below

# General Preprocessing

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.cluster import KMeans
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import warnings
warnings.filterwarnings('ignore')
```

```python
data_dir = Path(r'C:/Users/Guillem/Documents/GitHub/med_seagrass/data')
df_presence = pd.read_csv(data_dir / 'presence.txt', sep='\t')
df_absence = pd.read_csv(data_dir / 'absence.txt', sep='\t')
```

# Geographical variables preprocessing

```python
#  Determine geographical clusters to avoid spatial autocorrelation
# Step 1: Geographic clustering of coordinates using K-Means
print("="*80)
print("GEOGRAPHIC CLUSTERING (LATITUDE & LONGITUDE)")
print("="*80)

# Extract coordinates
coords = df_merge[['LONGITUDE', 'LATITUDE']].values

# Determine optimal number of clusters using the elbow method
inertias = []
K_range = range(3, 15)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(coords)
    inertias.append(kmeans.inertia_)

# Visualize elbow method
plt.figure(figsize=(10, 5))
plt.plot(K_range, inertias, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Number of Clusters (k)', fontsize=12)
plt.ylabel('Inertia (Within-Cluster Sum of Squares)', fontsize=12)
plt.title('Elbow Method to Determine Optimal K', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.show()

# Select k=8 (good balance between complexity and clustering)
# JUSTIFICATION: After the elbow, marginal benefit is low
optimal_k = 8
print(f"\n✅ Optimal number of clusters selected: {optimal_k}")

# Apply K-Means with optimal k
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df_merge['GEOGRAPHIC_ZONE'] = kmeans_final.fit_predict(coords)

print(f"\nDistribution of observations by geographic zone:")
print(df_merge['GEOGRAPHIC_ZONE'].value_counts().sort_index())

# Visualization of geographic zones
fig = plt.figure(figsize=(16, 10))
ax = plt.axes(projection=ccrs.PlateCarree())

# Add Mediterranean features (land and ocean first, as background)
ax.add_feature(cfeature.LAND, facecolor='wheat', alpha=0.4, zorder=1)
ax.add_feature(cfeature.OCEAN, facecolor='lightblue', alpha=0.3, zorder=1)
ax.add_feature(cfeature.BORDERS, linewidth=0.5, edgecolor='gray', linestyle=':', zorder=2)

# Define colors for each zone (using a qualitative colormap)
colors = plt.cm.tab10(np.linspace(0, 1, optimal_k))
zone_colors = {i: colors[i] for i in range(optimal_k)}

# Plot each geographic zone separately for better legend control
for zone in range(optimal_k):
    zone_data = df_merge[df_merge['GEOGRAPHIC_ZONE'] == zone]
    ax.scatter(zone_data['LONGITUDE'], zone_data['LATITUDE'], 
               c=[zone_colors[zone]], 
               label=f'Zone {zone}',
               s=60, alpha=0.8, edgecolors='black', linewidth=0.6,
               transform=ccrs.PlateCarree(), zorder=4)

# Add coastline OVER the points (higher zorder)
ax.add_feature(cfeature.COASTLINE, linewidth=1.2, edgecolor='black', zorder=5)

# Add categorical legend
legend = ax.legend(loc='upper left', title='Geographic Zones', 
                   frameon=True, fancybox=True, shadow=True,
                   fontsize=10, title_fontsize=12, ncol=2)
legend.set_zorder(6)  # Ensure legend is on top

# Set extent to Mediterranean region
ax.set_extent([-6, 43, 30, 46], crs=ccrs.PlateCarree())

# Add gridlines
gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--', zorder=3)
gl.top_labels = False
gl.right_labels = False

ax.set_xlabel('Longitude', fontsize=12)
ax.set_ylabel('Latitude', fontsize=12)
ax.set_title('Geographic Clustering of Observations (K-Means Clustering)', 
             fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print("\n✅ Variable GEOGRAPHIC_ZONE successfully created")
```

# Categorical variables preprocessing

```python
# Identify categorical variables
categorical_vars = df_merge.select_dtypes(include=['object']).columns.tolist()
print("Categorical variables:")
print(categorical_vars)
```

#### Category 1: BIO_CLASS

```python
bioclass_cat = df_merge["BIO_CLASS"].value_counts()
print("Unique BIO_CLASS categories:")
print(bioclass_cat)
```

```python
# Create a new column aggregating the different categories of BIO_CLASS into principal families
# (Zostera, Ruppia, Halophila, Cymodoceacea, Posidonia + Absence)
def categorize_bio_class(bio_class):
    bio_class_lower = str(bio_class).lower()
    
    if 'absence' in bio_class_lower:
        return 'Absence'
    elif 'zostera' in bio_class_lower:
        return 'Zostera'
    elif 'ruppia' in bio_class_lower:
        return 'Ruppia'
    elif 'halophila' in bio_class_lower:
        return 'Halophila'
    elif 'cymodocea' in bio_class_lower:
        return 'Cymodocea'
    elif 'posidonia' in bio_class_lower:
        return 'Posidonia'
    else:
        return 'Unknown'

df_merge['BIO_FAMILY'] = df_merge['BIO_CLASS'].apply(categorize_bio_class)

# Verify categorization
print("Distribution of BIO_FAMILY:")
bioclass_cat = df_merge['BIO_FAMILY'].value_counts()
print(bioclass_cat)

# Check for any Unknown classifications
unknown_count = (df_merge['BIO_FAMILY'] == 'Unknown').sum()
print(f"\nNumber of 'Unknown' classifications: {unknown_count}")

if unknown_count > 0:
    print("\nSamples with 'Unknown' BIO_FAMILY:")
    print(df_merge[df_merge['BIO_FAMILY'] == 'Unknown']['BIO_CLASS'].value_counts())

# Plot distribution
bioclass_cat.plot(kind='bar', figsize=(10,4))
plt.title('BIO_FAMILY Distribution')
plt.xlabel('Family')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Presence / Absence binary encoding as factor
df_merge['Presence'] = np.where(df_merge['BIO_CLASS'] == 'absence', False, True)
binary_cat = df_merge['Presence'].value_counts()
print("Binary counts (False = absence, True = presence):")
print(binary_cat)
```

```python
binary_cat.plot(kind='bar', figsize=(10,4), color=['orange', 'blue'], xlabel = "Class", ylabel="Frequency")
plt.title('Presence / Absence')
plt.xticks(ticks=[0, 1], labels=['presence', 'absence'])
plt.show()
```
