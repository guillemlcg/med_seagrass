## üó∫Ô∏è **Section 2.5: Geographic Zones & Spatial Distribution**

<div style="background-color: #fff3cd; padding: 15px; border-left: 4px solid #ffc107; margin: 10px 0;">
<strong style="color: #856404;">‚ö†Ô∏è Important Note:</strong> <span style="color: #000000;">To address potential <strong>spatial autocorrelation</strong> issues that could lead to overfitting in predictive models, the raw coordinates (LONGITUDE, LATITUDE) have been transformed into <strong>8 discrete geographic zones</strong> using K-Means clustering. This approach was implemented during the preprocessing phase following best practices for spatial modeling in ecological studies.</span>
</div>

<a id="zone-assignment"></a>

**Problem**: Using raw latitude/longitude as continuous predictors can cause models to:
- ‚úÖ **Memorize exact locations** rather than learn environmental patterns
- ‚úÖ **Overfit to spatial structure** in the training data
- ‚úÖ **Fail to generalize** to new geographic locations

**Solution**: K-Means clustering (k=8) applied to coordinates:
- ‚úÖ **Captures regional patterns** without overfitting to exact positions
- ‚úÖ **Reduces dimensionality** from 2 continuous ‚Üí 1 categorical variable
- ‚úÖ **Maintains spatial information** while improving model generalization

The following map visualizes the 8 geographic zones across the Mediterranean basin, showing how sampling points are clustered into distinct biogeographic regions:

```python
# Import required libraries for mapping
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Load the processed dataframe
df = pd.read_csv(r'C:/Users/Guillem/Documents/GitHub/med_seagrass/data/pres_abs_merge_def.csv')

# Create figure for geographic zones visualization
fig = plt.figure(figsize=(16, 10))
ax = plt.axes(projection=ccrs.PlateCarree())

# Add Mediterranean features (land and ocean first, as background)
ax.add_feature(cfeature.LAND, facecolor='wheat', alpha=0.4, zorder=1)
ax.add_feature(cfeature.OCEAN, facecolor='lightblue', alpha=0.3, zorder=1)
ax.add_feature(cfeature.BORDERS, linewidth=0.5, edgecolor='gray', linestyle=':', zorder=2)

# Define colors for each zone (using a qualitative colormap)
optimal_k = 8
colors = plt.cm.tab10(np.linspace(0, 1, optimal_k))
zone_colors = {i: colors[i] for i in range(optimal_k)}

# Plot each geographic zone separately for better legend control
for zone in range(optimal_k):
    zone_data = df[df['GEOGRAPHIC_ZONE'] == zone]
    ax.scatter(zone_data['LONGITUDE'], zone_data['LATITUDE'], 
               c=[zone_colors[zone]], 
               label=f'Zone {zone}',
               s=60, alpha=0.8, edgecolors='black', linewidth=0.6,
               transform=ccrs.PlateCarree(), zorder=4)

# Add coastline OVER the points (higher zorder)
ax.add_feature(cfeature.COASTLINE, linewidth=1.2, edgecolor='black', zorder=5)

# Add categorical legend
legend = ax.legend(loc='upper left', title='Geographic Zones', 
                   frameon=True, fancybox=True, shadow=True,
                   fontsize=10, title_fontsize=12, ncol=2)
legend.set_zorder(6)  # Ensure legend is on top

# Set extent to Mediterranean region
ax.set_extent([-6, 43, 30, 46], crs=ccrs.PlateCarree())

# Add gridlines
gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--', zorder=3)
gl.top_labels = False
gl.right_labels = False

ax.set_xlabel('Longitude', fontsize=12)
ax.set_ylabel('Latitude', fontsize=12)
ax.set_title('Geographic Clustering of Mediterranean Seagrass Observations\n(K-Means with k=8 zones)', 
             fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# Print zone distribution statistics
print("\nüìä DISTRIBUTION OF OBSERVATIONS BY GEOGRAPHIC ZONE:")
print("=" * 50)
zone_counts = df['GEOGRAPHIC_ZONE'].value_counts().sort_index()
for zone, count in zone_counts.items():
    percentage = (count / len(df)) * 100
    print(f"Zone {zone}: {count:>4} observations ({percentage:>5.1f}%)")
print("=" * 50)
print(f"Total:  {len(df):>4} observations (100.0%)")
```

#### **Implications for Modeling**

**Use of `GEOGRAPHIC_ZONE` vs. Raw Coordinates**:

| Approach | Advantages | Disadvantages | Recommendation |
|----------|------------|---------------|----------------|
| **Raw LONGITUDE/LATITUDE** | Maximum spatial precision | High risk of overfitting, memorizes exact locations | ‚ùå **Not recommended** for final models |
| **GEOGRAPHIC_ZONE (k=8)** | Better generalization, captures regional patterns | Loses within-zone spatial variation | ‚úÖ **Recommended** for Random Forest and tree-based models |
| **Both combined** | Best of both worlds | Increased dimensionality, potential redundancy | ‚ö†Ô∏è Use with caution, monitor for overfitting |

**Key Considerations**:
1. **For the Random Forest model** (as per Effrosynidis et al. 2018): Use `GEOGRAPHIC_ZONE` as a categorical predictor
2. **For spatial validation**: Consider stratified sampling by zone to ensure geographic representation
3. **For model interpretation**: Zone membership can reveal biogeographic patterns in seagrass distribution
4. **Alternative approaches**: Spatial cross-validation, geographic blocking, or distance-based features

> **üí° Best Practice**: When evaluating model performance, compare results using:
> - Model with `GEOGRAPHIC_ZONE` only
> - Model with raw coordinates only  
> - Model with both (if no overfitting detected)
>
> The version with `GEOGRAPHIC_ZONE` typically shows better generalization to unseen locations.

##### **4. Model Performance with Stratified Cross-Validation**

Evaluate model performance using 10-fold stratified cross-validation as per Effrosynidis et al. (2018).

```python
# Define evaluation metrics
scoring_metrics = {
    'accuracy': 'accuracy',
    'precision': make_scorer(precision_score, zero_division=0),
    'recall': make_scorer(recall_score),
    'f1': make_scorer(f1_score),
    'roc_auc': 'roc_auc'
}

# Initialize Random Forest model
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42,
    n_jobs=-1,
    class_weight='balanced'
)

# Perform 10-fold stratified cross-validation
print("üîÑ Performing 10-Fold Stratified Cross-Validation")
print("=" * 70)
print(f"   ‚Ä¢ Dataset: {len(X)} samples, {len(predictor_cols)} features")
print(f"   ‚Ä¢ Class distribution: {y.value_counts().to_dict()}")

cv_stratified = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
cv_results_stratified = cross_validate(
    rf_model, X, y, 
    cv=cv_stratified, 
    scoring=scoring_metrics,
    return_train_score=True,
    n_jobs=-1
)

print("\n‚úÖ Cross-validation completed!")
print("=" * 70)
```

```python
# Display cross-validation results
print("\nüìä STRATIFIED 10-FOLD CROSS-VALIDATION RESULTS:")
print("=" * 80)
print(f"{'Metric':<20} {'Train Mean':<15} {'Train Std':<15} {'Test Mean':<15} {'Test Std':<12}")
print("=" * 80)

results_stratified = {}
for metric in ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']:
    train_mean = cv_results_stratified[f'train_{metric}'].mean()
    train_std = cv_results_stratified[f'train_{metric}'].std()
    test_mean = cv_results_stratified[f'test_{metric}'].mean()
    test_std = cv_results_stratified[f'test_{metric}'].std()
    
    results_stratified[metric] = {
        'train_mean': train_mean,
        'train_std': train_std,
        'test_mean': test_mean,
        'test_std': test_std
    }
    
    print(f"{metric.upper():<20} {train_mean:>13.4f}  {train_std:>13.4f}  {test_mean:>13.4f}  {test_std:>10.4f}")

print("=" * 80)

# Compare with paper results
paper_accuracy = 0.934  # Effrosynidis et al. (2018) reported 93.4% accuracy
test_accuracy = results_stratified['accuracy']['test_mean']
diff = (test_accuracy - paper_accuracy) * 100

print(f"\nüìÑ COMPARISON WITH EFFROSYNIDIS ET AL. (2018):")
print(f"   ‚Ä¢ Paper reported accuracy: {paper_accuracy:.1%}")
print(f"   ‚Ä¢ Our model accuracy: {test_accuracy:.1%}")
print(f"   ‚Ä¢ Difference: {diff:+.2f} percentage points")

if abs(diff) < 2:
    print(f"   ‚úÖ Results are closely aligned with the reference study!")
elif diff > 0:
    print(f"   ‚ú® Our model slightly outperforms the reference study!")
else:
    print(f"   ‚ö†Ô∏è Our model underperforms compared to the reference study.")
```

##### **5. Spatial Cross-Validation (GroupKFold by Geographic Zone)**

To address potential **spatial autocorrelation** issues, we now perform cross-validation using **GroupKFold** with `GEOGRAPHIC_ZONE` as groups. This ensures that:
- ‚úÖ All samples from the same geographic zone stay together in either train or test fold
- ‚úÖ Model is tested on completely unseen geographic regions
- ‚úÖ More realistic assessment of generalization to new locations
- ‚úÖ Prevents spatial leakage between train and test sets

```python
# Perform Spatial Cross-Validation using GroupKFold
print("üó∫Ô∏è Performing Spatial Cross-Validation (GroupKFold by Geographic Zone)")
print("=" * 80)

# Get geographic zones as groups
groups = df['GEOGRAPHIC_ZONE']
n_zones = df['GEOGRAPHIC_ZONE'].nunique()

print(f"   ‚Ä¢ Total geographic zones: {n_zones}")
print(f"   ‚Ä¢ Samples per zone:")
for zone in sorted(df['GEOGRAPHIC_ZONE'].unique()):
    count = (groups == zone).sum()
    presence = ((groups == zone) & (y == True)).sum()
    absence = ((groups == zone) & (y == False)).sum()
    print(f"      - Zone {zone}: {count:>4} samples (Presence: {presence:>4}, Absence: {absence:>4})")

# Initialize GroupKFold (n_splits = number of zones)
cv_group = GroupKFold(n_splits=n_zones)
print(f"\n   ‚Ä¢ Cross-validation strategy: {n_zones}-fold GroupKFold")
print(f"   ‚Ä¢ Each fold tests on 1 complete geographic zone")

# Perform cross-validation
print(f"\nüîÑ Running spatial cross-validation...")
cv_results_group = cross_validate(
    rf_model, X, y, 
    cv=cv_group, 
    groups=groups,
    scoring=scoring_metrics,
    return_train_score=True,
    n_jobs=-1
)

print("\n‚úÖ Spatial cross-validation completed!")
print("=" * 80)
```

```python
# Display Spatial Cross-Validation Results
print("\nüìä SPATIAL CROSS-VALIDATION RESULTS (GroupKFold by Zone):")
print("=" * 80)
print(f"{'Metric':<20} {'Train Mean':<15} {'Train Std':<15} {'Test Mean':<15} {'Test Std':<12}")
print("=" * 80)

results_group = {}
for metric in ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']:
    train_mean = cv_results_group[f'train_{metric}'].mean()
    train_std = cv_results_group[f'train_{metric}'].std()
    test_mean = cv_results_group[f'test_{metric}'].mean()
    test_std = cv_results_group[f'test_{metric}'].std()
    
    results_group[metric] = {
        'train_mean': train_mean,
        'train_std': train_std,
        'test_mean': test_mean,
        'test_std': test_std
    }
    
    print(f"{metric.upper():<20} {train_mean:>13.4f}  {train_std:>13.4f}  {test_mean:>13.4f}  {test_std:>10.4f}")

print("=" * 80)
```

##### **6. Comparison: Stratified CV vs Spatial CV**

Compare the performance between standard stratified cross-validation and spatial cross-validation to assess the impact of spatial autocorrelation.

```python
# Compare Stratified vs Spatial Cross-Validation
print("\nüî¨ DETAILED COMPARISON: STRATIFIED CV vs SPATIAL CV")
print("=" * 90)
print(f"{'Metric':<15} {'Stratified':<12} {'Spatial':<12} {'Difference':<15} {'% Change':<12} {'Impact'}")
print("=" * 90)

comparison_data = []
for metric in ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']:
    strat_test = results_stratified[metric]['test_mean']
    group_test = results_group[metric]['test_mean']
    diff = group_test - strat_test
    pct_diff = (diff / strat_test) * 100
    
    # Determine impact level
    if abs(pct_diff) < 1:
        impact = "‚úÖ Negligible"
    elif abs(pct_diff) < 3:
        impact = "‚ö†Ô∏è Minor"
    elif abs(pct_diff) < 5:
        impact = "‚ö†Ô∏è Moderate"
    else:
        impact = "üî¥ Significant"
    
    comparison_data.append({
        'metric': metric,
        'stratified': strat_test,
        'spatial': group_test,
        'diff': diff,
        'pct_diff': pct_diff,
        'impact': impact
    })
    
    print(f"{metric.upper():<15} {strat_test:>10.4f}  {group_test:>10.4f}  {diff:>+13.4f}  {pct_diff:>+10.2f}%  {impact}")

print("=" * 90)

# Overall assessment
strat_accuracy = results_stratified['accuracy']['test_mean']
group_accuracy = results_group['accuracy']['test_mean']
accuracy_drop = (strat_accuracy - group_accuracy) * 100

print(f"\nüéØ KEY FINDINGS:")
print(f"   ‚Ä¢ Stratified CV Accuracy: {strat_accuracy:.1%}")
print(f"   ‚Ä¢ Spatial CV Accuracy: {group_accuracy:.1%}")
print(f"   ‚Ä¢ Accuracy Drop: {accuracy_drop:.2f} percentage points")

if accuracy_drop < 1:
    print(f"\n   ‚úÖ EXCELLENT: Minimal spatial autocorrelation detected!")
    print(f"      ‚Üí Model generalizes well across different geographic regions")
    print(f"      ‚Üí Spatial patterns are well-captured by environmental predictors")
elif accuracy_drop < 3:
    print(f"\n   ‚úÖ GOOD: Minor spatial autocorrelation detected")
    print(f"      ‚Üí Model shows good generalization with slight regional effects")
    print(f"      ‚Üí Geographic zone feature helps capture spatial patterns")
elif accuracy_drop < 5:
    print(f"\n   ‚ö†Ô∏è MODERATE: Some spatial autocorrelation present")
    print(f"      ‚Üí Model performance varies across geographic regions")
    print(f"      ‚Üí Consider region-specific feature engineering")
else:
    print(f"\n   üî¥ SIGNIFICANT: Strong spatial autocorrelation detected!")
    print(f"      ‚Üí Model is overfitting to spatial patterns in training data")
    print(f"      ‚Üí Stratified CV accuracy is optimistically biased")
    print(f"      ‚Üí Use spatial CV for realistic performance estimates")

# Compare with paper
paper_accuracy = 0.934
print(f"\nüìÑ COMPARISON WITH EFFROSYNIDIS ET AL. (2018):")
print(f"   ‚Ä¢ Paper reported: {paper_accuracy:.1%}")
print(f"   ‚Ä¢ Our Stratified CV: {strat_accuracy:.1%} ({(strat_accuracy - paper_accuracy)*100:+.2f} pp)")
print(f"   ‚Ä¢ Our Spatial CV: {group_accuracy:.1%} ({(group_accuracy - paper_accuracy)*100:+.2f} pp)")

if abs(group_accuracy - paper_accuracy) < abs(strat_accuracy - paper_accuracy):
    print(f"   ‚úÖ Spatial CV provides more comparable results to the paper!")
else:
    print(f"   ‚ÑπÔ∏è Stratified CV is closer to paper results (may not have used spatial CV)")

print("=" * 90)
```

```python
# Final Interpretation and Recommendations
print("\n" + "="*90)
print("üéØ FINAL ASSESSMENT & RECOMMENDATIONS")
print("="*90)

# Calculate key metrics
acc_drop = (strat_accuracy - group_accuracy) * 100
acc_drop_pct = (acc_drop / strat_accuracy) * 100
cv_coefficient_group = (group_acc.std() / group_acc.mean()) * 100
max_acc_group = group_acc.max()
min_acc_group = group_acc.min()
acc_range_group = max_acc_group - min_acc_group

print(f"\nüìà PERFORMANCE SUMMARY:")
print(f"   ‚Ä¢ Stratified 10-Fold CV: {strat_accuracy:.1%}")
print(f"   ‚Ä¢ Spatial {n_zones}-Fold CV:    {group_accuracy:.1%}")
print(f"   ‚Ä¢ Performance drop:     {acc_drop:.2f} percentage points ({acc_drop_pct:.2f}%)")
print(f"   ‚Ä¢ Paper benchmark:      {paper_accuracy:.1%}")

print(f"\nüîç SPATIAL AUTOCORRELATION ASSESSMENT:")
if acc_drop < 1:
    print(f"   ‚úÖ MINIMAL spatial autocorrelation (drop < 1%)")
    print(f"      ‚Üí Environmental predictors fully capture spatial patterns")
    print(f"      ‚Üí GEOGRAPHIC_ZONE feature is working effectively")
    spatial_status = "excellent"
elif acc_drop < 3:
    print(f"   ‚úÖ LOW spatial autocorrelation (drop 1-3%)")
    print(f"      ‚Üí Model generalizes well across regions")
    print(f"      ‚Üí Slight geographic effects are well-managed")
    spatial_status = "good"
elif acc_drop < 5:
    print(f"   ‚ö†Ô∏è MODERATE spatial autocorrelation (drop 3-5%)")
    print(f"      ‚Üí Some regional overfitting detected")
    print(f"      ‚Üí Consider additional spatial features")
    spatial_status = "moderate"
else:
    print(f"   üî¥ HIGH spatial autocorrelation (drop > 5%)")
    print(f"      ‚Üí Significant regional overfitting")
    print(f"      ‚Üí Stratified CV estimates are optimistically biased")
    spatial_status = "high"

print(f"\nüé≤ MODEL STABILITY:")
print(f"   ‚Ä¢ Fold variability (CV): {cv_coefficient_group:.2f}%")
print(f"   ‚Ä¢ Accuracy range: [{min_acc_group:.3f}, {max_acc_group:.3f}] (Œî={acc_range_group:.3f})")

if cv_coefficient_group < 2:
    print(f"   ‚úÖ EXCELLENT stability across geographic zones")
    stability_status = "excellent"
elif cv_coefficient_group < 3:
    print(f"   ‚úÖ GOOD stability with minor regional variation")
    stability_status = "good"
elif cv_coefficient_group < 5:
    print(f"   ‚ö†Ô∏è MODERATE stability - some zones harder to predict")
    stability_status = "moderate"
else:
    print(f"   üî¥ LOW stability - significant regional differences")
    stability_status = "low"

print(f"\nüìä COMPARISON WITH LITERATURE:")
gap = group_accuracy - paper_accuracy
if abs(gap) < 0.01:
    print(f"   ‚úÖ Results closely match Effrosynidis et al. (2018)")
    comparison_msg = "aligned"
elif gap > 0:
    print(f"   ‚ú® Model outperforms paper by {gap*100:.2f} percentage points")
    comparison_msg = "outperforming"
else:
    print(f"   ‚ö†Ô∏è Model underperforms paper by {abs(gap)*100:.2f} percentage points")
    comparison_msg = "underperforming"

print(f"\nüí° RECOMMENDATIONS FOR MODELING PHASE:")
print(f"   {'-'*85}")

if spatial_status in ['excellent', 'good']:
    print(f"   ‚úÖ 1. Use SPATIAL CV (GroupKFold) for final model evaluation")
    print(f"         ‚Üí More realistic estimate of generalization performance")
    print(f"         ‚Üí Better assessment for deployment to new regions")
    print(f"   ‚úÖ 2. Current feature set is well-designed")
    print(f"         ‚Üí GEOGRAPHIC_ZONE effectively captures spatial patterns")
    print(f"         ‚Üí Environmental predictors are informative")
else:
    print(f"   ‚ö†Ô∏è 1. MANDATORY: Use Spatial CV (GroupKFold) for evaluation")
    print(f"         ‚Üí Stratified CV significantly overestimates performance")
    print(f"   ‚ö†Ô∏è 2. Consider additional spatial feature engineering:")
    print(f"         ‚Üí Distance-based features (to known hotspots)")
    print(f"         ‚Üí Regional environmental interaction terms")
    print(f"         ‚Üí Spatial lag variables")

if stability_status in ['moderate', 'low']:
    print(f"   ‚ö†Ô∏è 3. Investigate zone-specific performance:")
    print(f"         ‚Üí Identify which zones have lower accuracy")
    print(f"         ‚Üí Consider ensemble approaches (region-specific models)")
    print(f"         ‚Üí Increase training data in challenging regions")

print(f"   üìã 4. Validation strategy for final model:")
print(f"         ‚Üí Primary: Spatial CV (GroupKFold by zone) - {group_accuracy:.1%}")
print(f"         ‚Üí Secondary: Stratified CV for comparison - {strat_accuracy:.1%}")
print(f"         ‚Üí Report both metrics with clear interpretation")

print(f"   üéØ 5. Model deployment considerations:")
print(f"         ‚Üí Expected accuracy on new regions: ~{group_accuracy:.1%}")
print(f"         ‚Üí Performance may vary by zone: {min_acc_group:.1%} - {max_acc_group:.1%}")
print(f"         ‚Üí Monitor performance when applying to unseen geographies")

print(f"\nüìù CONCLUSION:")
conclusions = []
if spatial_status in ['excellent', 'good']:
    conclusions.append("Model shows strong generalization across Mediterranean regions")
if stability_status in ['excellent', 'good']:
    conclusions.append("Predictions are stable across different geographic zones")
if comparison_msg == "aligned":
    conclusions.append("Performance aligns with published benchmarks")
elif comparison_msg == "outperforming":
    conclusions.append("Enhanced features and methodology improve upon published benchmarks")

if conclusions:
    for i, conclusion in enumerate(conclusions, 1):
        print(f"   {i}. {conclusion}")
else:
    print(f"   Model requires additional refinement for robust spatial generalization")

print(f"\n{'='*90}")
print(f"‚úÖ SPATIAL CROSS-VALIDATION ANALYSIS COMPLETE")
print(f"{'='*90}\n")
```

```python
# Import PyCaret
from pycaret.classification import *
import pandas as pd
import numpy as np

print("üîß PyCaret Classification Module Loaded")
print("=" * 70)
```

```python
# Prepare data for PyCaret
# Create a clean dataframe with predictors and target
pycaret_data = X.copy()
pycaret_data['Presence'] = y.values

# Convert boolean target to integer for PyCaret
pycaret_data['Presence'] = pycaret_data['Presence'].astype(int)

print("üìä Dataset prepared for PyCaret:")
print(f"   ‚Ä¢ Shape: {pycaret_data.shape}")
print(f"   ‚Ä¢ Features: {pycaret_data.shape[1] - 1}")
print(f"   ‚Ä¢ Target: Presence (0=Absence, 1=Presence)")
print(f"   ‚Ä¢ Class distribution:")
print(f"      - Class 0 (Absence): {(pycaret_data['Presence']==0).sum()} samples")
print(f"      - Class 1 (Presence): {(pycaret_data['Presence']==1).sum()} samples")
print("=" * 70)
```

#### **2. Model Evaluation with Stratified Cross-Validation**

Train and evaluate all models using stratified 10-fold cross-validation.

```python
# Setup PyCaret environment first (if not already done)
print("\nüîß Setting up PyCaret environment...")
print("=" * 70)

# Setup with stratified cross-validation
clf_setup = setup(
    data=pycaret_data,
    target='Presence',
    session_id=42,
    fold=10,
    fold_strategy='stratifiedkfold',
    verbose=False,
    html=False
)

print("‚úÖ PyCaret setup completed!")

# Train and evaluate specific models with Stratified CV
print("\nü§ñ Training models with STRATIFIED CV...")
print("=" * 70)

# Define models to compare (using correct PyCaret IDs)
models_to_compare = [
    'lr',      # Logistic Regression  
    'ridge',   # Ridge Classifier
    'lda',     # Linear Discriminant Analysis
    'svm',     # Support Vector Machine (Linear SVC equivalent)
    'knn',     # K-Nearest Neighbors
    'dt',      # Decision Tree
    'rf'       # Random Forest
]

# Model names for display
model_names = {
    'lr': 'Logistic Regression',
    'ridge': 'Ridge Classifier',
    'lda': 'Linear Discriminant Analysis',
    'svm': 'SVM - Linear Kernel',
    'knn': 'K Neighbors Classifier',
    'dt': 'Decision Tree Classifier',
    'rf': 'Random Forest Classifier'
}

print("\nüìã Models to be evaluated:")
for model_id in models_to_compare:
    print(f"   ‚Ä¢ {model_id}: {model_names[model_id]}")

# Create models
results_stratified_pycaret = []

for model_id in models_to_compare:
    print(f"\n   Training {model_names[model_id]}...")
    try:
        model = create_model(model_id, verbose=False, fold=10)
        
        # Get cross-validation results
        results = pull()
        
        # The results DataFrame structure in PyCaret:
        # - Index contains fold numbers (0, 1, 2, ..., Mean, Std)
        # - Columns contain metrics (Accuracy, AUC, Recall, Prec., F1, Kappa, MCC)
        # We want the 'Mean' row which contains average CV scores
        
        # Find the Mean row
        if 'Mean' in results.index:
            row_data = results.loc['Mean']
        else:
            # Fallback: use the second-to-last row
            row_data = results.iloc[-2]
        
        # Extract metrics with proper error handling
        results_stratified_pycaret.append({
            'Model': model_names[model_id],
            'Accuracy': float(row_data['Accuracy']) if 'Accuracy' in row_data else 0.0,
            'Precision': float(row_data['Prec.']) if 'Prec.' in row_data else 0.0,
            'Recall': float(row_data['Recall']) if 'Recall' in row_data else 0.0,
            'F1': float(row_data['F1']) if 'F1' in row_data else 0.0,
            'AUC': float(row_data['AUC']) if 'AUC' in row_data else 0.0
        })
        print(f"   ‚úÖ {model_names[model_id]} trained successfully")
        print(f"      Accuracy: {row_data['Accuracy']:.4f}, AUC: {row_data['AUC']:.4f}")
    except Exception as e:
        print(f"   ‚ùå Error training {model_names[model_id]}: {str(e)}")
        continue

# Create summary dataframe
stratified_results_df = pd.DataFrame(results_stratified_pycaret)

print("\n‚úÖ All models trained with Stratified CV!")
print(f"   Successfully trained {len(stratified_results_df)} models")
print("=" * 70)

# Display results
if len(stratified_results_df) > 0:
    print("\nüìä STRATIFIED CV RESULTS SUMMARY:")
    print("=" * 80)
    print(stratified_results_df.to_string(index=False))
    print("=" * 80)
```

```python
# Display Stratified CV results
print("\nüìä STRATIFIED CV RESULTS (10-Fold):")
print("=" * 90)
print(stratified_results_df.sort_values('Accuracy', ascending=False).to_string(index=False))
print("=" * 90)
```

### **5.2 Random Forest Variable Importance for Family Classification**

Compare multiple algorithms using both Stratified and Spatial cross-validation. Following the paper, we use **Macro F1-Score** as the primary evaluation metric due to class imbalance.

```python
# Prepare PyCaret data for family classification
from pycaret.classification import *

# Ensure GEOGRAPHIC_ZONE is in the dataset (if not already in predictor_cols)
pycaret_cols = predictor_cols.copy()
if 'GEOGRAPHIC_ZONE' not in pycaret_cols:
    pycaret_cols.append('GEOGRAPHIC_ZONE')

# Create dataset for PyCaret (with BIO_FAMILY as target)
pycaret_family = df_family[pycaret_cols + ['BIO_FAMILY']].copy()

print("üîß PYCARET SETUP FOR FAMILY CLASSIFICATION")
print("=" * 70)
print(f"   ‚Ä¢ Dataset shape: {pycaret_family.shape}")
print(f"   ‚Ä¢ Target variable: BIO_FAMILY (5 classes)")
print(f"   ‚Ä¢ Features: {len(pycaret_cols)}")
print(f"   ‚Ä¢ Evaluation metric: Macro F1-Score (class-balanced)")
print(f"   ‚Ä¢ Class distribution:")
for family, count in y_family.value_counts().items():
    print(f"      - {family}: {count} ({count/len(y_family)*100:.1f}%)")
print("=" * 70)
```

```python
# Setup 1: Stratified CV (10-fold)
print("\nüîÑ SETUP 1: STRATIFIED CROSS-VALIDATION")
print("=" * 80)

clf_family_stratified = setup(
    data=pycaret_family,
    target='BIO_FAMILY',
    fold=10,
    fold_strategy='stratifiedkfold',
    session_id=42,
    verbose=False,
    html=False,
    n_jobs=-1,
    fix_imbalance=False,  # Keep original distribution as per paper
    remove_outliers=False,
    normalize=False,
    transformation=False,
    pca=False,
    feature_selection=False
)

print("‚úÖ Stratified CV setup complete!")
```

```python
# Train models with Stratified CV
print("\nüéØ TRAINING MODELS WITH STRATIFIED CV")
print("=" * 80)

models_to_test_family = {
    'lr': 'Logistic Regression',
    'ridge': 'Ridge Classifier',
    'lda': 'Linear Discriminant Analysis',
    'svm': 'SVM - Linear Kernel',
    'knn': 'K Neighbors Classifier',
    'dt': 'Decision Tree Classifier',
    'rf': 'Random Forest Classifier'
}

results_stratified_family = []

for model_id, model_name in models_to_test_family.items():
    print(f"   ‚Ä¢ Training {model_name}...", end="")
    try:
        model = create_model(model_id, verbose=False, fold=10)
        cv_results = pull()
        
        # Get Mean row metrics
        if 'Mean' in cv_results.index:
            row_data = cv_results.loc['Mean']
        else:
            row_data = cv_results.iloc[-2]  # Second to last row is usually Mean
        
        results_stratified_family.append({
            'Model': model_name,
            'Model_ID': model_id,
            'Accuracy': float(row_data['Accuracy']),
            'Precision': float(row_data['Prec.']),
            'Recall': float(row_data['Recall']),
            'F1': float(row_data['F1']),
            'CV_Type': 'Stratified'
        })
        print(f" ‚úì (F1: {results_stratified_family[-1]['F1']:.4f})")
    except Exception as e:
        print(f" ‚úó Failed: {str(e)}")

stratified_results_family_df = pd.DataFrame(results_stratified_family)
print("\n‚úÖ Stratified CV training complete!")
print(stratified_results_family_df[['Model', 'F1']].to_string(index=False))
print("=" * 80)
```

#### **‚ö†Ô∏è CRITICAL: Reset PyCaret Environment**

Before training with spatial CV, we need to completely reset PyCaret's internal state to ensure the new cross-validation strategy is properly applied.

```python
# CRITICAL: Reset PyCaret environment before spatial CV
# This ensures the new CV strategy is properly applied
print("\nüîÑ RESETTING PYCARET ENVIRONMENT...")
print("=" * 80)

# Force garbage collection to clear memory
import gc
gc.collect()

# Re-import PyCaret to get fresh environment
from pycaret.classification import *

print("‚úÖ Environment reset complete!")
print("=" * 80)
```

```python
# Setup 2: Spatial CV (GroupKFold by zone) - FRESH SETUP
print("\nüó∫Ô∏è SETUP 2: SPATIAL CROSS-VALIDATION (GroupKFold)")
print("=" * 80)

# For spatial CV, we need to use fold_groups
clf_family_spatial = setup(
    data=pycaret_family,
    target='BIO_FAMILY',
    fold=8,  # 8 geographic zones
    fold_strategy='groupkfold',
    fold_groups='GEOGRAPHIC_ZONE',
    session_id=42,
    verbose=False,
    html=False,
    n_jobs=-1,
    fix_imbalance=False,
    remove_outliers=False,
    normalize=False,
    transformation=False,
    pca=False,
    feature_selection=False
)

print("‚úÖ Spatial CV setup complete!")
```

```python
# Train models with Spatial CV
print("\nüó∫Ô∏è TRAINING MODELS WITH SPATIAL CV (GroupKFold)")
print("=" * 80)

results_spatial_family = []

for model_id, model_name in models_to_test_family.items():
    print(f"   ‚Ä¢ Training {model_name}...", end="")
    try:
        model = create_model(model_id, verbose=False, fold=8)
        cv_results = pull()
        
        # Get Mean row metrics
        if 'Mean' in cv_results.index:
            row_data = cv_results.loc['Mean']
        else:
            row_data = cv_results.iloc[-2]  # Second to last row is usually Mean
        
        results_spatial_family.append({
            'Model': model_name,
            'Model_ID': model_id,
            'Accuracy': float(row_data['Accuracy']),
            'Precision': float(row_data['Prec.']),
            'Recall': float(row_data['Recall']),
            'F1': float(row_data['F1']),
            'CV_Type': 'Spatial'
        })
        print(f" ‚úì (F1: {results_spatial_family[-1]['F1']:.4f})")
    except Exception as e:
        print(f" ‚úó Failed: {str(e)}")

spatial_results_family_df = pd.DataFrame(results_spatial_family)
print("\n‚úÖ Spatial CV training complete!")
print(spatial_results_family_df[['Model', 'F1']].to_string(index=False))
print("=" * 80)
```

### **5.4 Results Comparison: Stratified vs Spatial CV**

Compare model performance across both cross-validation strategies and with the paper's reported results.

```python
# Combine and compare results
comparison_family_df = pd.merge(
    stratified_results_family_df,
    spatial_results_family_df,
    on=['Model', 'Model_ID'],
    suffixes=('_Stratified', '_Spatial')
)

# Calculate differences
comparison_family_df['F1_Difference'] = comparison_family_df['F1_Spatial'] - comparison_family_df['F1_Stratified']
comparison_family_df['F1_Drop_%'] = (comparison_family_df['F1_Difference'] / comparison_family_df['F1_Stratified']) * 100

# Sort by Spatial F1 (primary metric per paper)
comparison_family_df = comparison_family_df.sort_values('F1_Spatial', ascending=False)

print("\nüìä COMPREHENSIVE MODEL COMPARISON - FAMILY CLASSIFICATION")
print("=" * 120)
print(f"{'Model':<22} {'Stratified F1':<15} {'Spatial F1':<15} {'Difference':<15} {'Drop %':<10} {'Assessment'}")
print("=" * 120)

for _, row in comparison_family_df.iterrows():
    drop_pct = row['F1_Drop_%']
    if abs(drop_pct) < 5:
        assessment = "‚úÖ Stable"
    elif abs(drop_pct) < 10:
        assessment = "‚ö†Ô∏è Moderate"
    else:
        assessment = "üî¥ High Drop"
    
    print(f"{row['Model']:<22} {row['F1_Stratified']:>13.4f}  {row['F1_Spatial']:>13.4f}  {row['F1_Difference']:>+13.4f}  {drop_pct:>+8.1f}%  {assessment}")

print("=" * 120)

# Paper comparison
paper_f1_dt = 0.371  # Decision Tree with all features
paper_f1_rf = 0.397  # Random Forest with top-10% features (best in paper)

print(f"\nüìÑ COMPARISON WITH EFFROSYNIDIS ET AL. (2018) - Section 4.2:")
print("=" * 80)
print(f"   Paper Results (Macro F1):")
print(f"      ‚Ä¢ Decision Tree (all features): {paper_f1_dt:.3f}")
print(f"      ‚Ä¢ Random Forest (top-10% features): {paper_f1_rf:.3f} ‚≠ê Best in paper")

# Our best results
best_stratified_family = comparison_family_df.iloc[0]
best_f1_strat = best_stratified_family['F1_Stratified']
best_f1_spatial = best_stratified_family['F1_Spatial']

print(f"\n   Our Results (Macro F1):")
print(f"      ‚Ä¢ Best Stratified CV: {best_stratified_family['Model']} = {best_f1_strat:.3f}")
print(f"      ‚Ä¢ Best Spatial CV: {best_stratified_family['Model']} = {best_f1_spatial:.3f}")

# Compare with paper
if best_f1_spatial > paper_f1_rf:
    diff = (best_f1_spatial - paper_f1_rf) * 100
    print(f"\n   ‚ú® Our model outperforms paper by {diff:.1f} percentage points (Spatial CV)")
elif abs(best_f1_spatial - paper_f1_rf) < 0.02:
    print(f"\n   ‚úÖ Our results closely match the paper's best performance")
else:
    diff = (paper_f1_rf - best_f1_spatial) * 100
    print(f"\n   ‚ö†Ô∏è Paper's result is {diff:.1f} percentage points higher")

print("=" * 80)
```

```python
# VERIFICATION: Check if CV strategies produced different results
print("\nüîç VERIFICATION: Checking if CV strategies differ")
print("=" * 80)

max_f1_diff = comparison_family_df['F1_Difference'].abs().max()
max_acc_diff = abs(comparison_family_df['F1_Stratified'] - comparison_family_df['F1_Spatial']).max()

if max_f1_diff < 0.001:
    print("‚ö†Ô∏è WARNING: F1 scores are nearly identical between CV strategies!")
    print(f"   Maximum difference: {max_f1_diff:.6f}")
    print("\n   This suggests:")
    print("   1. Either the CV strategy is not being properly switched")
    print("   2. OR the models truly show identical performance (rare but possible)")
    print("\n   üí° Recommendation: If you see this warning consistently,")
    print("      consider restarting the kernel and re-running from the reset cell.")
else:
    print(f"‚úÖ SUCCESS: CV strategies produced different results")
    print(f"   Maximum F1 difference: {max_f1_diff:.4f}")
    print(f"   This confirms proper CV strategy switching")

print("=" * 80)
```

### **Spatial Autocorrelation Impact**

<div style="background-color: #fff3e0; padding: 20px; border-left: 6px solid #FF9800; margin: 15px 0; border-radius: 5px;">
<h4 style="margin-top: 0; color: #E65100;">‚ö†Ô∏è Critical Finding</h4>
<p style="color: #000000; line-height: 1.8; margin: 0;">
<strong>Spatial cross-validation reveals 3-10% lower performance</strong> than stratified CV, indicating models partially rely on spatial proximity rather than purely environmental features. This has important implications:
</p>
<ul style="color: #000000; line-height: 1.8;">
    <li><strong>Stratified CV:</strong> Overestimates performance due to spatial leakage between train/test</li>
    <li><strong>Spatial CV:</strong> Provides realistic generalization to new geographic regions</li>
    <li><strong>Production Deployment:</strong> Should expect performance closer to Spatial CV metrics</li>
</ul>
